```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data_mahasiswa = {
    'NIM': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Nama': ['Andi', 'Budi', 'Citra', 'Dewi', 'Eko'],
    'Nilai_Tugas': [80, 70, 90, 85, 60],
    'Nilai_UTS': [75, 65, 85, 80, 55],
    'Nilai_UAS': [90, 70, 88, 82, 50]
}
```

```python
df = pd.DataFrame(data_mahasiswa)
```

```python
df['Nilai_Akhir'] = df[['Nilai_Tugas', 'Nilai_UTS', 'Nilai_UAS']].mean(axis=1)
```

```python
df['Keterangan'] = df['Nilai_Akhir'].apply(lambda x: 'Lulus' if x >= 70 else 'Tidak Lulus')
```

```python
df.to_csv('nilai_mahasiswa.csv', index=False)
```

```python
plt.figure(figsize=(10, 6))
bar_width = 0.25
index = range(len(df))
```

```python
plt.bar([i - bar_width for i in index], df['Nilai_Tugas'], width=bar_width, label='Tugas', color='skyblue')
plt.bar(index, df['Nilai_UTS'], width=bar_width, label='UTS', color='orange')
plt.bar([i + bar_width for i in index], df['Nilai_UAS'], width=bar_width, label='UAS', color='green')
```

```python
plt.xlabel('Nama Mahasiswa')
plt.ylabel('Nilai')
plt.title('Perbandingan Nilai Tugas, UTS, dan UAS Mahasiswa')
plt.xticks(index, df['Nama'])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
